```python
# Importação
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Impontando a Base de Dados
df_videos = pd.read_excel('videosYT.xlsx', 'videos')
display(df_videos)
```

```python
# Gráfico de Dispersão
fig = sns.scatterplot(data=df_videos, x="Nº de Views", y="Nº de Likes", hue="Responsável", style="Responsável", palette=["red", "black"])
plt.show()
```

```python
# Gráfico de Dispersão(Relacional)
grafico_rel = sns.relplot(data=df_videos, x="Nº de Views", y="Nº de Likes", hue="Categoria", col="Responsável")
grafico_rel.set_titles("Responsável: {col_name}")
plt.show(grafico_rel)
```

```python
# Gráfico de Linha
df_inscritos = pd.read_excel("videosYT.xlsx", "Inscritos")

graf_linha = sns.lineplot(data=df_inscritos, x="Mês/Ano", y="Inscritos", color="red")
plt.show(graf_linha)
```

```python
# Criando Histogramas
fig = sns.displot(data=df_videos, x="Nº de Views", hue="Responsável", col="Responsável", rug=True)
plt.show(fig)
```

```python
# Criando uma Regressão Linear
graf_regre = sns.regplot(data=df_videos, x="Nº de Views", y="Nº de Likes")
plt.show(graf_regre)
```

```python
graf_regre = sns.lmplot(data=df_videos, x="Nº de Views", y="Nº de Likes", hue="Responsável", markers=['o', 'x'])
plt.show(graf_regre)
```
